```python
import re
```

```python
def clean(text):
    text = re.sub("\d:\d+:\d+\.\d S(\d+|\?): ", "", text)
    text = re.sub("\[.+?\]", "", text)
    text = re.sub("\-", " ", text)
    text = re.sub("\s{2,}", " ", text)
    new_text = text.strip()
    return new_text
    
```

```python
with open('/Users/TracyShields/Scribie/Elizabeth-Osder-Story-in-a-Bottle.txt') as fh:
    lines = fh.readlines()

lines = [clean(line) for line in lines]
lines = [line for line in lines if len(line.strip()) > 0]

raw_text = ' '.join(lines)
```

```python
clean(raw_text)
```
